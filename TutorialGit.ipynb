# Tutorial de uso básico do GitHub

Este tutorial tem como objetivo orientar o usuário para que este possa utilizar os recursos do GitHub, famoso serviço de Web Hosting para compartilhar softwares open-source, e publicar seus trabalhos de modo simples. 

## Habilidades a serem desenvolvidas:

- Criação e clonagem de repositórios
- Submeter arquivos e mudanças remotamente (via terminal Linux)
- Download de arquivos separados (_single-file download_) 
- Edição do README.md

## Como funciona o GitHub?

O GitHub é um serviço, acima de tudo, para compartilhamento de dados. É como um Google Drive onde todos os usuários houvessem permitido que seus arquivos fossem baixados por outras pessoas. O interessante é que você pode manusear seus repositórios (local onde se armazena um software, para este poder ser recuperado e instalado) via terminal ou pelo próprio site. 

### Passo 1: Cadastramento (_signing-up_)
- Acesse https://github.com/join?source=header-home
- Preencha os campos com o nome de usuário que deseja, email e senha.
- Clique em "Continue":
<img src="./tutorial_git_figs/tutorialgit.jpg">
- Ao final, na última tela, clique em "Start a project" e receberá um e-mail de confirmação. A opção "Read the guide" é opcional, e oferece um tutorial bem interativo para operar no próprio site, ao invés do terminal. 

### Passo 2: Criando o repositório
Quando clicar no email de confirmação, uma aba será aberta com as mesmas opções de antes, e clicando em "Start a project" você será direcionado para a criação de um repositório:
<img src="./tutorial_git_figs/repo.jpg"/>
- Atenção: MARQUE A OPÇÃO "Initialize this repository with a README"
- Uma vez que o diretório está criado, vamos começar a trabalhar no terminal.
### Passo 3: Clonando o diretório

Para que possamos utilizar os comandos ```git``` no terminal, precisamos instalar a ferramenta via linha de comando. Basta usar:
```bash
sudo apt-get install git
```
Com a instalação concluída, ou a ferramenta já estando instalada, vamos clonar o diretório. 
- Copie a URL do repositório clicando em "Clone or download":
<img src="./tutorial_git_figs/clone.jpg">
- Digite o comando no terminal, substituindo 'endereço' pela URL:
```bash
git clone endereço
```
<img src="./tutorial_git_figs/clone2.png">

Agora já existe um diretório com o mesmo nome do repositório na pasta em que você executou o comando. 

### Submetendo arquivos
- Entre no diretório 
- Crie um arquivo de texto entitulado como "teste.txt" 
```bash
cd repositorio-teste/
touch teste.txt
ls ```

O comando ```touch``` cria o arquivo e o ```ls``` exibe o conteúdo do diretório.

Para adicionar os arquivos ao diretório, use:
```bash
git add -A
```
A opção -A serve para adicionar todos. Use ```git add --help``` para mais opções.

Agora, para validar (_comittar_) suas alterações, use:
```bash
git commit -m "att"
``` 
A mensagem após ```-m``` é de sua preferência, mas sempre uso "att". Lembre-se das aspas.

Como você provavelmente nunca usou esse comando, o ```git```pede uma identificação, apenas. Então use:
```bash
git config --global user.email "seuemail@mail.com"
git config --global user.name "seu nome"
```
Normalmente eu uso o email da minha conta GitHub e meu usuário no lugar do nome. 

Agora reexecute o comando ```commit``` como mostrado acima.

Por fim, é necessária uma autenticação com seu usuário (NÃO É O EMAIL) e a senha do github. Use:
```bash
git push origin master
``` 
O resultado no terminal deve ser algo como (o comando ```git pull``` é desnecessário):
<img src="./tutorial_git_figs/clone3.png">

Como podemos ver, o arquivo de texto "teste.txt" foi adicionado ao GitHub com sucesso. Atualize a página (F5) antes de conferir. 
<img src="./tutorial_git_figs/clone4.png">

## Editando o README.md

Essa parte é meramente estética, mas bem útil e simples para postagem de links importantes. Na lista de arquivos, clique em README.md. Aparecerá uma janela nova, onde iremos fazer as edições. Esse arquivo é responsável pela apresentação do repositório. Clique no lápis ao lado de "History", como mostra a figura:
<img src="./tutorial_git_figs/readme.png">
Uma janela com o código fonte será aberta, e nela faremos as edições, que podem ser visualizadas clicando em "Preview changes".
<img src="./tutorial_git_figs/readme2.png">

- Para títulos, digita-se o texto ao lado de um #.
- Para subseções, digita-se o texto ao lado de ##
- Links: _*_ [legenda do link] (link) 
- Imagens: _!_[legenda da imagem (que nem aparece)] (url da imagem)
- Retire os espaços na entre o '!', os colchetes [] e a url 
- No link, só há um espaço, que é entre o asterisco '*' e a legenda
- Resultado:

<img src="./tutorial_git_figs/readme3.png">
