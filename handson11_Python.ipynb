## Códigos em Python
Podemos gerar um canal Rayleigh de duas maneiras. A primeira é pela forma abordada acima e a segunda é basicamente a norma entre as duas componentes de W.





```python
import numpy as np
import matplotlib.pyplot as plt

N=20000                                       # Número de amostras
x=np.arange(0,5,0.1)                          # Eixo x
sigma=1                                       # Parâmetro Rayleigh
u=np.random.rand(N)                           # Amostras aleatórias
## Montagem do canal
r=sigma*np.sqrt(-2*np.log(u))                 # Dados aleatorios da distribuição Rayleigh
r_ac= x/(sigma**2)*np.exp(-(x/sigma)**2/2)    # Rayleigh PDF

## plotting
plt.figure()
plt.subplot(2,1,1)
plt.hist(r,x,color='midnightblue')
plt.axis([0,5,0,1500])
plt.title('Histograma para N=20000 amostras')
plt.subplot(2,1,2)
plt.plot(x,r_ac)
plt.title('Rayleigh PDF')

plt.tight_layout()
plt.show()
```
